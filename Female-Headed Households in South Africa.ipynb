```python
df = pd.concat([train, test])
df.head()
```

```python
df['indivisual_per_household'] = df['total_individuals']/df['total_households']

df.head()
```

```python
df['WARD'] = df['ward'].apply(lambda x: str(x)[10:])
df['PCODE'] = df['ADM4_PCODE'].apply(lambda x: str(x)[2:])
df['ADM4'] = df['ADM4_PCODE'].apply(lambda x: str(x)[0:2])
```

```python
df = df.drop(['ADM4_PCODE', 'ward'], axis = 1)
```

```python
df_dum = pd.get_dummies(df.drop('target', axis = 1), drop_first = True)
df_dum.head()
```

```python
#list(dummies.columns)
```

```python
df1 = pd.concat([df_dum,df['target']], axis = 1)
df1.head()
```

```python
train_c = df1[pd.notnull(df1['target'])]
test_c = df1[pd.isnull(df1['target'])]
```

```python
X = train_c.drop('target', axis = 1)
y = train_c['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y.astype(float))
```

```python
# Using this to test a model on 5 different splits
kf = KFold(n_splits=5, shuffle=False)

scores = []
for train, test in kf.split(train_c):
    rf = RandomForestRegressor(n_estimators=100)
    rf.fit(X.iloc[train], y.iloc[train])
    rmse = sqrt(mean_squared_error(y.iloc[test], rf.predict(X.iloc[test])))
    scores.append(rmse)
    print(rmse)

print("Average score in 5-fold CV:", np.mean(scores))
```

```python
# Using this to test a model on 5 different splits
kf = KFold(n_splits=5, shuffle=True)

scores = []
for train, test in kf.split(train_c):
    rf = RandomForestRegressor(n_estimators=100)
    rf.fit(X.iloc[train], y.iloc[train])
    rmse = sqrt(mean_squared_error(y.iloc[test], rf.predict(X.iloc[test])))
    scores.append(rmse)
    print(rmse)

print("Average score in 5-fold CV:", np.mean(scores))
```

```python
kf = KFold(n_splits=5, shuffle=False)
for md in [3, 4, 5, 7, 9, 12, 15]:
    scores = []
    for train, test in kf.split(train_c):
        rf = RandomForestRegressor(n_estimators=100, max_depth=md)
        rf.fit(X.iloc[train], y.iloc[train])
        rmse = sqrt(mean_squared_error(y.iloc[test], rf.predict(X.iloc[test])))
        scores.append(rmse)
    print("Average score in 5-fold CV:", np.mean(scores), "for max depth:", md)
```

```python
y = y.astype(float)
for md in [3, 4, 5, 7, 9, 12, 15]:
    scores = []
    for train, test in kf.split(train_c):
        rf = CatBoostRegressor(n_estimators=100, max_depth=md)
        rf.fit(X.iloc[train], y.iloc[train])
        rmse = sqrt(mean_squared_error(y.iloc[test], rf.predict(X.iloc[test])))
        scores.append(rmse)
    print("Average score in 5-fold CV:", np.mean(scores), "for max depth:", md)
```
